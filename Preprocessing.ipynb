```python
redCardOnly = rawInc.copy()
drop_red_list = []

count_red_home = 0
for red_card_home in redCardOnly['Red_Card_Home']:
    if red_card_home > 0:
        drop_red_list.append(count_red_home)
    count_red_home = count_red_home + 1
    
count_red_away = 0
for red_card_away in redCardOnly['Red_Card_Away']:
    if red_card_away > 0 and count_red_away not in drop_red_list:
        drop_red_list.append(count_red_away)
    count_red_away = count_red_away + 1
```

```python
def rangeTwo(x, y):
    Final = []
    while x < y:
        Final.append(x)
        x = x + 1
    return Final
    
redCardData = minuteByMinute.copy()
maxRange = int(len(redCardData))
actual_drop_list = []

for drop_row in rangeTwo(0, maxRange):
    if drop_row not in drop_red_list:
        actual_drop_list.append(drop_row)

redCardData = redCardData.drop(actual_drop_list)
redCardOnly = redCardOnly.drop(actual_drop_list)
```

```python
redCardData
```

```python
primary_red_list = []
for check_red in redCardData.index:
    secondary_red_list = []
    count = 0
    for minute_count in redCardData.loc[check_red]:
        minute_count_list = minute_count.split(',')
        for individual_action in minute_count_list:
            if individual_action == 'Red_Card_Home' or individual_action == 'Red_Card_Away':
                secondary_red_list.append(count - 2)
        count = count + 1
    primary_red_list.append(secondary_red_list)
```

```python
count_unusable = 0
drop_unusable_data = []
for unusable_data in primary_red_list:
    if len(unusable_data) == 2:
        drop_unusable_data.append(count_unusable)
    count_unusable = count_unusable + 1
```

```python
drop_index_list = []
for drop_unusable_real in drop_unusable_data:
    drop_index = int(redCardData.iloc[drop_unusable_real].name)
    drop_index_list.append(drop_index)
    
redCardData = redCardData.drop(drop_index_list)
redCardData
```

```python
primary_double_list = []
for check_double in redCardData.index:
    checker_for_double = False
    for double_count in redCardData.loc[check_double]:
        double_count_list = double_count.split(',')
        if len(double_count_list) > 1:
            checker_for_double = True
    if checker_for_double == True:
        primary_double_list.append(check_double)
```

```python
redCardData = redCardData.drop(primary_double_list)
```

```python
double_red_list = []
for last_drop in redCardData.index:
    red_one_check = 0
    for last_red_check in redCardData.loc[last_drop]:
        if last_red_check == 'Red_Card_Home' or last_red_check == 'Red_Card_Away':
            red_one_check = red_one_check + 1
    if red_one_check > 1:
        double_red_list.append(last_drop)
```

```python
redCardData = redCardData.drop(double_red_list)
```

```python
# Nothing = 0
# Goal Home = 1
# Yellow Home = 2
# Red Card Home = 3
# Own Home = 4
# Goal Away = -1
# Yellow Away = -2
# Red Card Away = -3
# Own Away = -4
# 3.566 x 70% = 2.489 / 1.067

beforeRedCard = redCardData.copy()
afterRedCard = redCardData.copy()

beforeRedCard = beforeRedCard.drop(['Unnamed: 0', 'ID'], axis=1)
afterRedCard = afterRedCard.drop(['Unnamed: 0', 'ID'], axis=1)
```

```python
for transform in beforeRedCard.index:
    for transform_column in rangeTwo(0, len(beforeRedCard.columns)):
        name_column = 'M' + str(transform_column)
        if beforeRedCard.loc[transform, name_column] == '-':
            beforeRedCard.loc[transform, name_column] = 0
        elif beforeRedCard.loc[transform, name_column] == 'Goal_Home':
            beforeRedCard.loc[transform, name_column] = 1
        elif beforeRedCard.loc[transform, name_column] == 'Yellow_Home':
            beforeRedCard.loc[transform, name_column] = 2
        elif beforeRedCard.loc[transform, name_column] == 'Red_Card_Home':
            beforeRedCard.loc[transform, name_column] = 3
        elif beforeRedCard.loc[transform, name_column] == 'Own_Home':
            beforeRedCard.loc[transform, name_column] = 4
        elif beforeRedCard.loc[transform, name_column] == 'Goal_Away':
            beforeRedCard.loc[transform, name_column] = -1
        elif beforeRedCard.loc[transform, name_column] == 'Yellow_Away':
            beforeRedCard.loc[transform, name_column] = -2
        elif beforeRedCard.loc[transform, name_column] == 'Red_Card_Away':
            beforeRedCard.loc[transform, name_column] = -3
        elif beforeRedCard.loc[transform, name_column] == 'Own_Away':
            beforeRedCard.loc[transform, name_column] = -4

beforeRedCard
```

```python
for remove_after in beforeRedCard.index:
    red_detect = False
    for column_after in rangeTwo(0, len(beforeRedCard.columns)):
        name_before = 'M' + str(column_after)
        if beforeRedCard.loc[remove_after, name_before] == 3 or beforeRedCard.loc[remove_after, name_before] == -3:
            red_detect = True
        elif red_detect == True:
            beforeRedCard.loc[remove_after, name_before] = 0

beforeRedCard
```

```python
for transform_after in afterRedCard.index:
    for transform_column_after in rangeTwo(0, len(afterRedCard.columns)):
        name_column_after = 'M' + str(transform_column_after)
        if afterRedCard.loc[transform_after, name_column_after] == '-':
            afterRedCard.loc[transform_after, name_column_after] = 0
        elif afterRedCard.loc[transform_after, name_column_after] == 'Goal_Home':
            afterRedCard.loc[transform_after, name_column_after] = 1
        elif afterRedCard.loc[transform_after, name_column_after] == 'Yellow_Home':
            afterRedCard.loc[transform_after, name_column_after] = 2
        elif afterRedCard.loc[transform_after, name_column_after] == 'Red_Card_Home':
            afterRedCard.loc[transform_after, name_column_after] = 3
        elif afterRedCard.loc[transform_after, name_column_after] == 'Own_Home':
            afterRedCard.loc[transform_after, name_column_after] = 4
        elif afterRedCard.loc[transform_after, name_column_after] == 'Goal_Away':
            afterRedCard.loc[transform_after, name_column_after] = -1
        elif afterRedCard.loc[transform_after, name_column_after] == 'Yellow_Away':
            afterRedCard.loc[transform_after, name_column_after] = -2
        elif afterRedCard.loc[transform_after, name_column_after] == 'Red_Card_Away':
            afterRedCard.loc[transform_after, name_column_after] = -3
        elif afterRedCard.loc[transform_after, name_column_after] == 'Own_Away':
            afterRedCard.loc[transform_after, name_column_after] = -4

afterRedCard
```

```python
def rangeReverse(x, y):
    Final = []
    while y > x:
        y = y - 1
        Final.append(y)
    return Final

for remove_before in afterRedCard.index:
    red_detect_before = False
    for column_before in rangeReverse(0, len(afterRedCard.columns)):
        name_after = 'M' + str(column_before)
        if afterRedCard.loc[remove_before, name_after] == 3 or afterRedCard.loc[remove_before, name_after] == -3:
            red_detect_before = True
        elif red_detect_before == True:
            afterRedCard.loc[remove_before, name_after] = 0

afterRedCard
```
